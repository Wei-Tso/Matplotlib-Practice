```python
import matplotlib.pyplot as plt
```

## Method 1 : subplot2grid

```python
plt.figure()

ax1 = plt.subplot2grid((3,3) , (0 , 0) , colspan = 3 , rowspan = 1)
"""
(3,3) : 全部有 3 行 3 列 的小圖
(0 , 0) : 列行的索引值，皆從 0 開始，從 (0 , 0) 開始
colspan = 3 : 將 column1、column2、column3 的小圖合成一張大圖
rowspan = 1
"""
ax1.plot([1 , 2] , [1 , 2])
ax1.set_title("ax1 title")
"""
ax1.set_title() : 之前是用 plt.title，因為這裡是 axis，不是 figure，所以就用 ax.set_title()
ax.set_xlabel <=> plt.xlabel
ax.set_xlim <=> plt.xlim
"""

plt.tight_layout()
plt.show()
```

```python
plt.figure()

ax1 = plt.subplot2grid((3,3) , (0 , 0) , colspan = 3 , rowspan = 1)
ax1.plot([1 , 2] , [1 , 2])
ax1.set_title("ax1 title")

ax2 = plt.subplot2grid((3,3) , (1 , 0) , colspan = 2 , rowspan = 1)
"""
(1 , 0) : 列行的索引值，皆從 0 開始，從 第 1 列、第 0 行 開始
"""
ax2.plot([1 , 2] , [1 , 2])
ax2.set_title("ax2 title")

ax3 = plt.subplot2grid((3,3) , (1 , 2) , colspan = 1 , rowspan = 2)
"""
(1 , 2) : 從 第 1 列、第 2 行 開始
"""
ax4 = plt.subplot2grid((3,3) , (2 , 0) , colspan = 1 , rowspan = 1)
"""
(2 , 0) : 從 第 2 列、第 0 行 開始
"""
ax5 = plt.subplot2grid((3,3) , (2 , 1) , colspan = 1 , rowspan = 1)
"""
(2 , 1) : 從 第 2 列、第 1 行 開始
"""
plt.tight_layout()
plt.show()
```

## Method 2 : gridspec

```python
import matplotlib.gridspec as gridspec
```

```python
plt.figure()

gs = gridspec.GridSpec(3,3)
"""
(3,3) : 全部有 3 行 3 列
"""

ax1 = plt.subplot(gs[0 , :])
"""
gs[0 , :] : 在 gs(3,3) 中，小圖所佔的位置，皆以索引表示
gs[0 , :] : 第 0 列，第 0 ~ 2 行(共佔 3 行)
"""

ax2 = plt.subplot(gs[1 , :2])
"""
gs[1 , :2] : 第 1 列，第 0 ~ 1 行(共佔 2 行)
"""

ax3 = plt.subplot(gs[1: , 2])
"""
gs[1: , 2] : 第 1 ~ 2 列(共佔 2 列)，第 2 行
"""

ax4 = plt.subplot(gs[-1 , 0])
"""
gs[-1 , 0] : 倒數 第 1 列，第 0 行
"""

ax5 = plt.subplot(gs[-1 , -2])
"""
gs[-1 , -2] : 倒數 第 1 列，倒數 第 2 行
"""

plt.tight_layout()
plt.show()
```
