```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1 , 1 , 50)
y1 = 2 * x +1
y2 = x ** 2
```

```python
plt.figure()
plt.plot(x , y1)

plt.figure(num=3 , figsize=(8,5))
plt.plot(x , y2)
plt.plot(x , y1 , color='red' , linewidth=10.0 , linestyle='--')
"""
color='red' ：線的顏色
linewidth=1.0 ：線的寬度
linestyle='--' ：線的樣式
"""

plt.show()
```
