```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
```

```python
# 柱狀圖 plt.bar()
plt.bar(X , Y1)
plt.bar(X , -Y2)

# 去除 x , y 的ticks
plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

```python
plt.bar(X , Y1 , facecolor = "#9999ff" , edgecolor = "white")
plt.bar(X , -Y2 , facecolor = "#ff9999" , edgecolor = "white")
"""
facecolor：柱狀圖顏色
edgecolor：邊框顏色
"""

plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

```python
plt.bar(X , Y1 , facecolor = "#9999ff" , edgecolor = "white")
plt.bar(X , -Y2 , facecolor = "#ff9999" , edgecolor = "white")

for x , y in zip(X , Y1):
    plt.text(x , y , "%.2f"%y , ha = "center" , va = "bottom")
"""
x + 0.4：顯示位置
y + 0.05：顯示位置
ha：horizontal alignment
ha = "center"：水平對齊：居中
va = "bottom"：垂直對齊：
"""

for x , y in zip(X , Y2):
    plt.text(x , -y - 0.05 , "-%.2f"%y , ha = "center" , va = "top")
    
plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```
