```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]
```

```python
fig = plt.figure(figsize=(8 , 8))

left=0.1
bottom=0.1
width=0.8
height=0.8

ax1 = fig.add_axes([left , bottom , width , height])
ax1.plot(x , y , "r")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Title")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8 , 8))

left=0.1     # 距離 左側 為 0.1倍 的長度 (0.1 倍 的 figure 寬度)
bottom=0.1   # 距離 下側 為 0.1倍 的長度 (0.1 倍 的 figure 高度)
width=0.8    # 此圖的寬度為原 figure 的 0.8 倍
height=0.8   # 此圖的高度為原 figure 的 0.8 倍

ax1 = fig.add_axes([left , bottom , width , height])
ax1.plot(x , y , "r")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Title")

left=0.2
bottom=0.6
width=0.25
height=0.25

ax2 = fig.add_axes([left , bottom , width , height])
ax2.plot(x , y , "b")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("Title inside 1")

left=0
bottom=0
width=0.2
height=0.2

ax2 = fig.add_axes([left , bottom , width , height])
ax2.plot(x , y , "g")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("Figure origin point")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8 , 8))

left=0.1     # 距離 左側 為 0.1倍 的長度 (0.1 倍 的 figure 寬度)
bottom=0.1   # 距離 下側 為 0.1倍 的長度 (0.1 倍 的 figure 高度)
width=0.8    # 此圖的寬度為原 figure 的 0.8 倍
height=0.8   # 此圖的高度為原 figure 的 0.8 倍

ax1 = fig.add_axes([left , bottom , width , height])
ax1.plot(x , y , "r")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Title")

left=0.2
bottom=0.6
width=0.25
height=0.25

ax2 = fig.add_axes([left , bottom , width , height])
ax2.plot(x , y , "b")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("Title inside 1")

plt.axes([0.6 , 0.2 , 0.25 , 0.25])
plt.plot(y[::-1] , x , "g")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Title inside 2")

plt.tight_layout()
plt.show()
```
