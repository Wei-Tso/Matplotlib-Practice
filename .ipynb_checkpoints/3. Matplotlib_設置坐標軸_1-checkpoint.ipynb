```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1 , 1 , 50)
y1 = 2 * x +1
y2 = x ** 2
```

```python
plt.figure()
plt.plot(x , y1)

plt.figure()
plt.plot(x , y2)
plt.plot(x , y1 , color='red' , linewidth=1.0 , linestyle='--')

plt.xlim((-1 , 2))
plt.ylim((-2 , 3))
"""
xlim((-1 , 2)) ：x 軸的範圍
ylim((-2 , 3)) ：y 軸的範圍
"""

plt.xlabel("I am X")
plt.ylabel("I am Y")
"""
xlabel() ：x 軸的名稱
ylabel() ：y 軸的名稱
"""

plt.show()
```

```python
plt.figure()
plt.plot(x , y2)
plt.plot(x , y1 , color='red' , linewidth=1.0 , linestyle='--')

plt.figure()
plt.plot(x , y2)
plt.plot(x , y1 , color='red' , linewidth=1.0 , linestyle='--')

new_ticks = np.linspace(-1 , 2 , 5)
print(new_ticks)
plt.xticks(new_ticks)
"""
改變 x 軸顯示的座標刻度
"""
plt.yticks([-1 , 0 , 1 , 2 , 3] , ["really bad" , "bad" , "normal" , "good" , "really good"])
plt.xticks(new_ticks)
"""
改變 y 軸顯示的文字
plt.yticks([] , []) ：[] 內為一一對應
"""

plt.show()
```
