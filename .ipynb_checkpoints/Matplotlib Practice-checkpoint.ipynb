```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 100 個從 -5 ~ 2 的數值
x = np.linspace(-5 , 2 , 100)

y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10
y4 = x**2
```

```python
len(x)
```

```python
plt.plot(x , y1)
plt.show()
```

```python
plt.plot(x , y2)
plt.show()
```

```python
plt.plot(x , y3)
plt.show()
```

```python
plt.plot(x , y4)
```

```python
fig , ax = plt.subplots(2 , 3 , figsize=(8,5))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color="blue", label="y(x)") 
ax.plot(x, y2, color="red", label="y'(x)")
ax.plot(x, y3, color="green", label="y''(x)")
ax.set_xlabel("x") # x軸
ax.set_ylabel("y") # y軸
ax.legend(); # 顯示
```
