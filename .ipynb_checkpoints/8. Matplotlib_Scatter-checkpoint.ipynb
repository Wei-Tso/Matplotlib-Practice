```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1024
X = np.random.normal(0 , 1 , n)
Y = np.random.normal(0 , 1 , n)
T = np.arctan2(Y , X)
```

```python
plt.scatter(X , Y , s = 75 , c = T , alpha = 0.5)
plt.xlim(-1.5 , 1.5)
plt.ylim(-1.5 , 1.5)

plt.show()
```

```python
plt.scatter(X , Y , s = 75 , c = T , alpha = 0.5)
plt.xlim(-1.5 , 1.5)
plt.ylim(-1.5 , 1.5)

plt.xticks(())
plt.yticks(())
"""
.xticks(()) ：隱藏 x 軸的數值
.yticks(()) ：隱藏 y 軸的數值
"""

plt.show()
```

```python
plt.scatter(X , Y , s = 10 , c = T , alpha = 0.5)
"""
s ：點的大小
"""
plt.xlim(-1.5 , 1.5)
plt.ylim(-1.5 , 1.5)
plt.xticks(())
plt.yticks(())


plt.show()
```
