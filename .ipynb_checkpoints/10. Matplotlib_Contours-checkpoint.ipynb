```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# the height function
def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)
```

```python
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
```

```python
# 將 x , y 綁定為 網格X , Y
X , Y = np.meshgrid(x , y)

# 等高線圖 plt.contourf()
plt.contourf(X , Y , f(X , Y) , 8 , alpha = 0.75 , cmap = plt.cm.hot)
# plt.contourf(X , Y , f(X , Y) , 8 , alpha = 0.75 , cmap = plt.cm.cool)
"""
X：mesh 後的 x
Y：mesh 後的 y
f(X , Y)
8：(看以下範例)
"""

plt.show()
```

```python
X , Y = np.meshgrid(x , y)

plt.contourf(X , Y , f(X , Y) , 8 , alpha = 0.75 , cmap = plt.cm.cool)
C = plt.contour(X , Y , f(X , Y) , 8 , colors = "black" , linewidth = 0.5)

plt.show()
```
