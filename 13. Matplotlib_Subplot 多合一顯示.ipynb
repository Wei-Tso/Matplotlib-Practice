```python
import matplotlib.pyplot as plt
```

```python
plt.figure()

plt.subplot(2 , 3 , 1)
plt.plot([0,1] , [0,1])
"""
plt.subplot(2 , 3 , 1)：2 行，3 列，第 1 個位置
plt.plot([0,1] , [0,1])：(0,0) 至 (1,1) 的直線
"""

plt.subplot(2 , 3 , 2)
plt.plot([0,1] , [0,2])

plt.subplot(2 , 3 , 3)
plt.plot([0,1] , [0,3])

plt.subplot(2 , 3 , 4)
plt.plot([0,1] , [0,4])

plt.subplot(2 , 3 , 5)
plt.plot([0,1] , [0,5])

plt.subplot(2 , 3 , 6)
plt.plot([0,1] , [0,6])

plt.show()
```
